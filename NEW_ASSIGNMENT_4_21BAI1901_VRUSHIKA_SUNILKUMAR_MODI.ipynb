```python
y = df['quality'].apply(lambda y_value:1 if y_value>=7 else 0)
y
```

```python
X =df.drop(columns =['quality'],axis =1)
X.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scale =MinMaxScaler()
```

```python
X_scaled= pd.DataFrame(scale.fit_transform(X),columns =X.columns)
X_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.3,random_state=10)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

**LINEAR REGRESSION**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_predict =lr.predict(x_test)
```

```python
y_predict1=lr.predict(x_train)
```

```python
quality = pd.DataFrame({'Actual quality':y_test,'Predicted quality':y_predict})
quality
```

```python
from sklearn import metrics
```

```python
print("Testing Accuracy:",metrics.r2_score(y_test,y_predict))
print("Training Accuracy:",metrics.r2_score(y_train,y_predict1))
```

**LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
accuracy_score(y_test,pred)
```

```python
pd.crosstab(y_test,pred)
```
